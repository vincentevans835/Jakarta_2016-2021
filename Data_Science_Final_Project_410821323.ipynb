```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

jakarta19 = pd.read_excel('jakarta.xlsx')
jakarta16 = pd.read_excel('jakarta 16-18.xlsx')
```

```python
jakarta16
```

```python
jakarta19
```

```python
jakarta16 = jakarta16.dropna()
jakarta19 = jakarta19.dropna()
```

```python
jakarta16
```

```python
jakarta19
```

```python
jakarta = pd.merge(jakarta16,jakarta19,how='inner',on='Kelompok Umur')
jakarta = jakarta[0:14]
```

```python
jakarta
```

```python
outside = ['Age','Male','Male','Male','Female','Female','Female','Total','Total','Total','Male','Male','Male','Female','Female','Female','Total','Total','Total']
inside = ['','2016','2017','2018','2016','2017','2018','2016','2017','2018','2019','2020','2021','2019','2020','2021','2019','2020','2021']
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
jakarta.columns = hier_index
jakarta.set_index("Age",drop=True,inplace=True)
```

```python
jakarta = jakarta.astype(int)
```

```python
jakarta
```

```python
jakarta['Male'].plot(title='Male')
plt.savefig('Male.jpg')
```

```python
jakarta['Female'].plot(title='Female')
plt.savefig('Female.jpg')
```

```python
jakarta['Total'].plot(title='Total')
plt.savefig('Total.jpg')
```
